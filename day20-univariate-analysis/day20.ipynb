# 2. Numerical Data

## a. Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=5)
```

```python
#sns.distplot(df['Age'])
#sns.histplot(df['Age'], kde=True)
sns.histplot(
    df["Age"], kde=True,
    stat="density", kde_kws=dict(cut=3),
    alpha=.4, edgecolor=(1, 1, 1, .4),
)
```
